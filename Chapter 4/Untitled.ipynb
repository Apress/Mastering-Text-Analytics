```python
import nltk
from nltk import CFG
from nltk import Tree

# Define a simple CFG
grammar = CFG.fromstring("""
  S -> NP VP
  NP -> DT NN
  VP -> V NP
  DT -> 'the'
  NN -> 'cat' | 'dog'
  V -> 'chases' | 'sees'
""")

# Create a parser with the grammar
parser = nltk.ChartParser(grammar)

# Sentence to parse
sentence = 'the cat chases the dog'.split()

# Parse the sentence
for tree in parser.parse(sentence):
    print("Syntax Tree:")
    tree.pretty_print()
    
    # Optionally, draw the tree
    tree.draw()
```
