- Python RegEx for cleaning of text strings.

```python
import re

# Below is a sample text string, we found in a Tweet. 
sample_text = "Hi!!! I am John. Welcome @my house. You can get more info @ www.sk.com"

# ReEx to remove special characters and URLs from the sample_text.
usable_text = re.sub(r"[!@#%&*:]", "", sample_text)  # Remove special characters
usable_text = re.sub(r"http\S+|www\.\S+", "", usable_text)  # Clean of URLs.

print(usable_text)
```
